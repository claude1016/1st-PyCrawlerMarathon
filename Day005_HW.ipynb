```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests as re
import json 
r=re.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r_j=json.loads(r.text)

print('一次傳回:',len(r_j),'筆資料')
print('每一筆包含',len(r_j[1].keys()),'欄位')
print(r_j[0].keys())
print(r_j[1].keys())
print(r_j)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''
import pandas as pd
title=[]
createdAt=[]
commentCount=[]
likeCount=[]
for n in r_j:
    
    title.append(n['title'])
    createdAt.append(n['createdAt'])
    commentCount.append(n['commentCount'])
    likeCount.append(n['likeCount'])
data=pd.DataFrame({'title':title,'createdAt':createdAt,'commentCount':commentCount,'likeCount':likeCount})  
print(data)
```
