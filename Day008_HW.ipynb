### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'

'''
Your Code Here
'''
headers={'user-agent':'000'}
r = requests.get(url,headers=headers)
print(type(r.text))
r.encoding = 'utf-8'
bf_1=BeautifulSoup(r.text,'html5lib')
print(bf_1)
# print(r.text[0:600])
```
